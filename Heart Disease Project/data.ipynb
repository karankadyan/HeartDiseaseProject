Importing Libraries: Import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

Reading Data: Read the dataset 'framingham.csv' into a Pandas DataFrame.

```python
df = pd.read_csv('framingham.csv')
```

Displaying Data: Display the first few rows of the DataFrame to get an overview of the data.

```python
df.head()
```

```python

df.tail()
```

Data Information: Display information about the DataFrame, including data types and non-null counts.

```python
df.info()
```

Checking for Missing Values: Check for missing values in the DataFrame using the isnull() function.



```python
df.isnull()
```

Renaming Columns: Rename the 'male' column to 'sex' using the rename() function.

```python
df.rename(columns={'male':'sex'}, inplace = True)
```

Correlation Heatmap: Create a heatmap to visualize the correlation between columns in the DataFrame.

```python
plt.figure(figsize=(17,6))
sns.heatmap(df.corr(),annot=True)
```

Target Variable Distribution: Display the distribution of the 'TenYearCHD' target variable using a countplot.

```python
df['TenYearCHD'].value_counts()
```

```python
sns.countplot(x='TenYearCHD',data=df)
```

Missing Value Percentage: Calculate and display the percentage of missing values for each column.

```python
pct_missing = round((df.isnull().sum())/(df.isnull().count())*100,1)
pct_missing.sort_values(ascending = False).head()
```

Missing Value Heatmap: Create a heatmap to visualize the missing values in the DataFrame.

```python
sns.heatmap(data=df.isnull())
```

Distribution of Gender: Display the distribution of the 'sex' column using a countplot

```python
df['sex'].value_counts()
```

```python
sns.countplot(x='sex',data=df)
```

Removing Rows with Missing Values: Remove rows with missing values using the dropna() function.

```python
df.dropna(inplace=True)
```

Removing Duplicate Rows: Remove duplicate rows from the DataFrame using the drop_duplicates() function.

```python
df.drop_duplicates(inplace=True)
```

Saving Cleaned Data: Save the cleaned DataFrame to a new CSV file named 'cleaned_data.csv'.

```python
df.to_csv('cleaned_data.csv', index=False)
```

Loading Cleaned Data: Read the cleaned data from the CSV file.

Visualization of Missing Values: Create a heatmap to visualize the missing values in the cleaned data.



```python
clean = pd.read_csv('cleaned_data.csv')
sns.heatmap(data=clean.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
